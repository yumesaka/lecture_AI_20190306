```python
#몬테카를로 기법
# pi는 r=1인 원의 넒이와 같음.
```

```python
n = 10000
x = np.random.rand(n)
y = np.random.rand(n)

((x**2 + y**2) < 1).mean()*4

```

```python
plt.figure(figsize= (6,6))
xs = np.linspace(0,1,100)
ys = np.sqrt(1 - xs*xs)
plt.scatter(x,y)
plt.plot(xs, ys, c='r')
```

```python
(1 - xs ** xs)**0.5
```

```python
z = 0
n = 30 #주사위의 갯수
for i in range(n):
    x = np.random.randint(1,7,100000)
    z = z+x
plt.hist(z, bins= n *5+1, width = 0.6)
plt.show()
```
